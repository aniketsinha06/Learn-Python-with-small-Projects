### How to convert an image to a cartoon painting using OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#import the image
img = cv2.imread('scene1.jpg')
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
plt.imshow(image)
plt.grid
```

```python
#OpenCV imports images in BGR format. Convert the image from BGR to GRAY
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 5)
plt.imshow(gray)
```

```python
#Finding edges with adaptive threshold mean
edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 5, 5)
edges1 = cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)
plt.imshow(edges1)
```

```python
#Adding color to the outline we created
color = cv2.bilateralFilter(img, 9, 250, 250)
cartoon = cv2.bitwise_and(color, color, mask = edges)
```

```python
#Display the painting
cv2.imshow("Image", img)
cv2.imshow("Cartoon", cartoon)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Didplay using matplotlib
cartoon_plt = cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)
plt.imshow(cartoon_plt)
```
