```python
import cv2, time, pandas 

def detect_motion(count):
    static_back = None
    motion_list = [ None, None ]
  
    df = pandas.DataFrame(columns = ["Start", "End"]) 
    video = cv2.VideoCapture(0) 
    while True:
    
        motion = 0
        
        motion_flag = 0
    
        check, frame = video.read()

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
  
        gray = cv2.GaussianBlur(gray, (21, 21), 0) 
  
        if static_back is None: 
            static_back = gray 
            continue
  
        diff_frame = cv2.absdiff(static_back, gray) 
  
        thresh_frame = cv2.threshold(diff_frame, 30, 255, cv2.THRESH_BINARY)[1] 
        thresh_frame = cv2.dilate(thresh_frame, None, iterations = 2) 
  
        cnts,_ = cv2.findContours(thresh_frame.copy(),  
                       cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
  
        for contour in cnts: 
            if cv2.contourArea(contour) < 10000:
                continue
            motion = 1
  
        
        motion_list.append(motion) 
  
        motion_list = motion_list[-2:] 
    
        print(motion_list)
  
        if motion_list[-1] == 1 and motion_list[-2] == 0: 
            motion_flag = 1 
  
        if motion_list[-1] == 0 and motion_list[-2] == 1: 
            motion_flag = 1
        
        if motion == 1:
            print('motion detected')
        else:
            print('No motion')
        cv2.imshow("Threshold Frame", thresh_frame)
            
        count = count + 1
        key = cv2.waitKey(1) 
        if key == ord('q') or count == 300:
            break 

    video.release() 
    cv2.destroyAllWindows()
```

```python
import cv2

class Motion_Detector():
    def __init__(self, first_frame = None, motion = 0, count = 0):
        self.first_frame = first_frame
        self.motion = motion
        self.count = count
    
    def detect_motion(self):
        video = cv2.VideoCapture(0) 
        while True:
            
            self.motion = 0    
            check, frame = video.read()

            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
  
            gray = cv2.GaussianBlur(gray, (21, 21), 0) 
  
            if self.first_frame is None: 
                self.first_frame = gray 
                continue
  
            diff_frame = cv2.absdiff(self.first_frame, gray) 
  
            thresh_frame = cv2.threshold(diff_frame, 30, 255, cv2.THRESH_BINARY)[1] 
            thresh_frame = cv2.dilate(thresh_frame, None, iterations = 2) 
  
            cnts,_ = cv2.findContours(thresh_frame.copy(),  
                       cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 
  
            for contour in cnts: 
                if cv2.contourArea(contour) < 10000:
                    continue
                self.motion = 1
  
            if self.motion == 1:
                print('motion detected')
            else:
                print('No motion')
            cv2.imshow("Threshold Frame", thresh_frame)
            
            self.count = self.count + 1
            key = cv2.waitKey(1) 
            if key == ord('q') or self.count == 300:
                break 

        video.release() 
        cv2.destroyAllWindows()

if __name__=="__main__":
    detect = Motion_Detector(count = 300)
    detect.detect_motion()      
```
